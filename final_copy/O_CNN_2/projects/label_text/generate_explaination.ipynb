```python
def get_labels_explaination(labels, N=10):
    response = openai.Completion.create(
        model="text-davinci-003",
        prompt=[f"explain what is {label} in 20 words" for label in labels],
        max_tokens=30,
        n=N
    )
    explaination = [ch['text'].strip().lower() for ch in response['choices']]
    result = [[]]
    for i, expr in enumerate(explaination):
        result[-1].append(expr)
        if (i+1) % N == 0:
            result.append([])
    return result
```

```python
from constant import CLASS_LABELS_200
from tqdm import tqdm
from collections import defaultdict
import time
result = defaultdict(lambda : [])
tmp_labels = []

start = time.time()
for label in tqdm(CLASS_LABELS_200):
    tmp_labels.append(label)
    if len(tmp_labels) == 10:
        tmp_result = get_labels_explaination(tmp_labels)
        for label, explains in zip(tmp_labels, tmp_result):
            result[label] = explains
        tmp_labels = []
    if len(result) == 40:
        time.sleep(50)
```
